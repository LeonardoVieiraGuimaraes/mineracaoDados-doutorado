```python
import numpy as np
import os
import pandas as pd
```

```python
past = './DataSet/datatran'
```

```python
frames = []
inicio = 2017
fim = 2022
col = ['municipio', 'causa_acidente', 'tipo_acidente', 'classificacao_acidente', 'fase_dia', 'sentido_via', 'condicao_metereologica', 'tipo_pista', 'tracado_via', 'uso_solo']
for file in range(inicio,fim):
    print(past + str(file) + '.csv')
    df = pd.read_csv(past + str(file) + '.csv', delimiter = ';',  low_memory=False, encoding='UTF-8', index_col = False)
    df = df.drop(columns=['id'])
    frames.append(df)
```

```python
df1 = pd.DataFrame(pd.concat(frames,ignore_index = True, join="inner"))
df1 = df1.dropna()
```

```python
col = df1.columns
for i in col:
    indexn = df1[df1[i] == "(Null)"].index
    df1.drop(indexn , inplace=True)
    indexn = df1[df1[i] == "(null)"].index
    df1.drop(indexn , inplace=True)
```

```python
df1['km'] = df1['km'].replace(',','.', regex=True)
df1['longitude']= df1['longitude'].replace(',','.', regex=True)
df1['latitude']= df1['latitude'].replace(',','.', regex=True)
df1['km'] = pd.to_numeric(df1['km'], downcast='integer')
df1['br'] = pd.to_numeric(df1['br'], downcast='integer')
```

```python
df1 = df1.convert_dtypes()
df1.dtypes
```

```python
for i in df1.columns:
    if (df1[i].dtypes == 'string' and i != 'uf'):
        df1[i] = df1[i].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
        df1[i] = df1[i].str.title()
```

```python
df1.dtypes
```

```python
df1 = df1.convert_dtypes()
df1.dtypes
```

```python
df1['data_inversa'] = pd.to_datetime(df1['data_inversa'])
df1['ano'] = df1['data_inversa'].dt.year
df1['mes'] = df1['data_inversa'].dt.month_name()
df1['dia'] = df1['data_inversa'].dt.day
df1['dia_semana'] = df1['data_inversa'].dt.day_name()
```

```python
df1.loc[df1['mortos'] > 0, 'mortes'] = 1
df1.loc[df1['mortos'] == 0, 'mortes'] = 0
```

```python
df1 = df1.convert_dtypes()
df1.dtypes
```

```python
df1.to_csv('DataSet/Final/DataSet' + str(inicio) + '_' + str(fim-1) + '.csv', index=False, sep = ';', encoding="UTF-8")
```
