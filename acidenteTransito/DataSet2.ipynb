```python
import numpy as np
import os
import pandas as pd
```

```python
past = './DataSet/datatran'
```

```python
frames = []
inicio = 2017
fim = 2022
for file in range(inicio,fim):
    print(past + str(file) + '.csv')
    df = pd.read_csv(past + str(file) + '.csv', delimiter = ';',  low_memory=False, encoding='UTF-8', index_col = False)
    df = df.drop(columns=['id'])
    frames.append(df)
```

```python
df1 = pd.DataFrame(pd.concat(frames))
df1
```

```python
df1['data_inversa'] = pd.to_datetime(df1['data_inversa'])
df1['ano'] = df1['data_inversa'].dt.year
df1['mes'] = df1['data_inversa'].dt.month_name()
df1['dia'] = df1['data_inversa'].dt.day
df1['dia_semana'] = df1['data_inversa'].dt.day_name()
df1['km'] = pd.to_numeric(df1['km'], downcast='integer', errors = 'coerce')
df1['br'] = pd.to_numeric(df1['br'], downcast='float', errors = 'coerce')
df1['horario'] = pd.to_datetime(df1['horario']).dt.time
```

```python
df1.dtypes
```

```python
df1n = df1.convert_dtypes()
df1n.dtypes
```

```python
df1n
```

```python
df1n.columns.values
col = ['municipio', 'causa_acidente', 'tipo_acidente', 'classificacao_acidente', 'fase_dia', 'sentido_via', 'condicao_metereologica', 'tipo_pista', 'tracado_via', 'uso_solo', 'mes']
```

```python
for i in col:
    df1n[i] = df1n[i].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    df1n[i] = df1n[i].str.title()
    
```
