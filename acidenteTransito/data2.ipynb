```python
import numpy as np
import os
import pandas as pd
import chardet
```

```python
past = './DataSet/datatran'
```

```python
frames = []
inicio = 2007
fim = 2022
for file in range(inicio,fim):
    print(past + str(file) + '.csv')
    df = pd.read_csv(past + str(file) + '.csv', delimiter = ';',  low_memory=False)
    df = df.drop(columns=['id'])
    with open(past + str(file) + '.csv', 'rb') as rawdata:
        result = chardet.detect(rawdata.read(10000))

    # check what the character encoding might be
    print(result)
    if (file >= 2017):
        df = df.drop(columns=['latitude', 'longitude', 'regional','delegacia', 'uop'])
    frames.append(df)
```

```python
df1 = pd.DataFrame(pd.concat(frames))
```

```python
df1['data_inversa'] = pd.to_datetime(df1['data_inversa'])

df1['ano'] = df1['data_inversa'].dt.year
df1['mes'] = df1['data_inversa'].dt.month_name()
df1['dia'] = df1['data_inversa'].dt.day
df1['dia_semana'] = df1['data_inversa'].dt.day_name()
df1['km'] = pd.to_numeric(df1['km'], downcast='integer', errors = 'coerce')
df1['br'] = pd.to_numeric(df1['br'], downcast='float', errors = 'coerce')
```
